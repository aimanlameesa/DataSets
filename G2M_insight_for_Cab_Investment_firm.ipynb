Hypothesis Investigation

1. Most of the customers of pink cab are female. In other words, pink cabs are popular among female customers.

```python
plt.figure(figsize = (10, 10))
sns.countplot(x ='Gender', hue = "Company", data = new_dataset2)

plt.show()
```

It seems that yellow cabs are famous among both male and female customers.

2. Yellow cabs are popular for all age groups of customers.

```python
plt.figure(figsize = (15, 10))
sns.countplot(x ='Age', hue = "Company", data = new_dataset2)

plt.show()
```

It seems that our hypothesis is correct, that, yellow cabs are popular for all age groups of customers.

3. The popularity of the two cab companies can be different for different seasons.

```python
plt.figure(figsize = (30, 10))
sns.countplot(x ='Date of Travel', hue = "Company", data = new_dataset2)

plt.show()
```

It seems that yellow cab is popular for most of the time. Pink cab is famous for a few times during the time period.

4. The charged price to the customers are high for New York City, Washington DC and San Diego.

```python
import plotly.express as px

fig = px.bar(new_dataset2, x="City", y="Price Charged", color="Company")
fig.show()
```

The charged price is the highest for New York city and the second highest for Washington DC, but lower in San Diego than most of the other cities.

5. The cost of trip is similar for the cities as price charged.

```python
import plotly.express as px

fig = px.bar(new_dataset2, x="City", y="Cost of Trip", color="Company")
fig.show()
```

It seems that New York city has the highest cost of travel as the price charged from the customers. However, Chicage has a signicant high cost of trip compared with charged price from the cab customers. Chicage has the second highest cost of trip. Moreover, Los Angeles and Boston have higher cost of trip than most of the others.
